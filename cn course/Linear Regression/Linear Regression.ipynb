```python
import numpy as np
```

```python
data = np.loadtxt('data.csv', delimiter=',')
```

```python
data.shape
# rows , cols
```

```python
x = data[:,0].reshape(-1, 1)
y = data[:,1]
```

```python
x.shape, y.shape
```

```python
y
```

```python
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
x_train, x_test , y_train, y_test = model_selection.train_test_split(x,y)
```

```python
x_train.shape
```

```python
algo = LinearRegression()
algo.fit(x_train, y_train);
# gives error because we gave 1D array (one feature)
# but it required 2D array (2 features)

# solution is use .reshape(-1, 1)
```

```python
# algo creates y = mx+c
# m
m = algo.coef_[0]
m
```

```python
# c
c = algo.intercept_
c
```

```python
import matplotlib.pyplot as plt
```

```python
x_line = np.arange(30,70,0.1)
y_line = m * x_line + c
```

```python
plt.plot(x_line, y_line, c='r')
training1d = x_train.reshape(75)
plt.scatter(training1d, y_train)

plt.show()
```

```python
plt.plot(x_line, y_line, c='r')

training1d = x_test.reshape(25)
plt.scatter(x_test, y_test)

plt.show() 
```

```python
score_test = algo.score(x_test, y_test)
score_training = algo.score(x_train, y_train)
score_test, score_training
```
