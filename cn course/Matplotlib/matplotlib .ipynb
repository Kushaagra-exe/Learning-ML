```python
import pandas as pd
import csv
import collections
import matplotlib.pyplot as plt

amazon = pd.read_csv('amazon_jobs_dataset.csv')
```

```python
data ={}
with open('amazon_jobs_dataset.csv', encoding='UTF-8') as file:
    file_data = csv.DictReader(file)
    
    
    for row in file_data:
        date = row['Posting_date'].split()
        year = date[2]
        if year in data:
            data[year]+=1
        else:
            data[year]=1
        
data = dict(sorted(data.items()))

for i,j in data.items():
    print(f'{i} - {j}')
```

```python
plt.plot(data.keys(), data.values())
```

```python
month_data = {}
with open('amazon_jobs_dataset.csv', encoding = 'UTF-8') as file:
    file_data = csv.DictReader(file)
    
    for row in file_data:
        date = row['Posting_date'].split()
        month = date[0]
        
        if month in month_data:
            month_data[month]+=1
        else:
            month_data[month]=1

print(month_data)
```

```python
plt.bar(month_data.keys(), month_data.values())
plt.xticks(rotation=40)
plt.show()

for i,j in month_data.items():
    print(f'{i} - {j}')
```

```python
indian_jobs = {}
with open('amazon_jobs_dataset.csv', encoding='UTF-8') as file:
    file_data = csv.DictReader(file)
    
    for row in file_data:
        location = row['location'].split(', ')
        country = location[0]
        if(country=='IN'):
            city = location[2]
            
            if city in indian_jobs:
                indian_jobs[city]+=1
            else:
                indian_jobs[city]=1
                
print(indian_jobs)
```

```python
plt.pie(indian_jobs.values(), labels = indian_jobs.keys(), autopct = '%.2f%%')

plt.show()
```

```python
t = sum(indian_jobs.values())
for i,j in indian_jobs.items():
    pct = (j/t)*100
    print(f'{i} - {round(pct,2)}')
```

```python
java_jobs = {}
with open('amazon_jobs_dataset.csv', encoding='UTF-8') as file:
    file_data = csv.DictReader(file)
    
    for row in file_data:
        qualify = row['BASIC QUALIFICATIONS']
        if 'Java' in qualify or 'java'in qualify:
            date = row['Posting_date'].split()
            year = date[2]
            if year in java_jobs:
                java_jobs[year] +=1
            else:
                java_jobs[year]=1
                
java_jobs = dict(sorted(java_jobs.items()))
print(java_jobs)
```
