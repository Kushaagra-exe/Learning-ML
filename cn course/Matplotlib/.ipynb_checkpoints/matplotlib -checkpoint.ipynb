### plotting
    .scatter(x, y) -> points
    .plot(x2, y2, "g") -> colour - green
    .plot(x2, y2, "g--") -> dashed line
    # types of lines -, --, +, ^, p, 
    .plot(x2, y2,color="green",marker="--",linewidth=0.5)
    
### Titles
    plt.ylabel("x^3")
    plt.xlabel("x")
    plt.title("demo")
    plt.text(x, y, text)
     
### Additional params for graph
    plt.axis([0,10,0,200]) -> [xstart,xend,ystart,yend]
    plt.grid() lines in graph
    plt.lengend() -> mark the info of plots

```python
x= np.arange(0,5,0.1)
y = x**3
y2 = x**2
plt.plot(x, y, color="black", linewidth=5, label="x^3")
plt.plot(x, y2, color="blue", linewidth=5, label="x^2")

# plt.axis([0,10,0,200])

plt.ylabel("y")
plt.xlabel("x")
plt.legend()
plt.grid()
plt.text(2,60,"text")
plt.title("demo")
plt.show()
```

# BUBBLE CHART

```python
employees = [61,71,79,91,93,89,90,94,99,128,118,114,124,131]
year = [2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
revenue = [39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36]
```

```python
plt.scatter(year, revenue, s = 50, c='r', alpha =0.3) 
# s =size
# alpha = opaque-ness
```

```python
plt.scatter(year, revenue, s = employees) 
```

```python
c = np.arange(len(year))
plt.scatter(year, revenue, s = employees, c=c, alpha =0.3, marker='^') 
```

```python
plt.scatter(year, revenue, s = employees, c=c, alpha =0.3, edgecolor = 'blue' )
plt.text(year[0]+0.1,revenue[0]+0.4, employees[0])
plt.text(year[4]+0.1,revenue[4]+0.4, employees[4])

plt.show()
```

```python
c = np.arange(len(year))
plt.scatter(year, revenue, s = employees, c=c, alpha =0.3, marker='^') 
```

# Pie chart

```python
sizes = [3, 4, 6 ,2]
colors = ['blue', 'purple','red', 'pink']
labels = ['a', 'b', 'c', 'd']
```

```python
plt.pie(sizes, colors = colors, labels = labels, autopct = '%.2f%%')
plt.axis("equal")


plt.title("SPlit")
plt.show()
```

```python
explode = [0.1,0,0,0]


plt.pie(sizes, colors = colors, labels = labels, autopct = '%.2f%%', explode = explode, counterclock = False, startangle = 180)
plt.axis("equal")

plt.title("SPlit")
plt.show()
```
